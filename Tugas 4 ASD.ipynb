```python
Stack = ['Siput']

# Perintah

def Perintah () :
    Perintah = ["Perintah : " , "1. Push","2. Pop", "3. Top",
    "4. Size" , "5. Empty" , "6. Tampilkan stack" , "7. Keluar"]
    for i in Perintah :
        print(i)
    
    Opsi = input("Pilih menu : ")   
    if Opsi == "1" :
         Push (Stack)
    elif Opsi == "2" :
         Pop (Stack)
    elif Opsi == "3" :
         Top (Stack)                  
    elif Opsi == "4" :
         Size (Stack)
    elif Opsi == "5" :
         Empty (Stack)
    elif Opsi == "6" :
         Print (Stack)
    elif Opsi == "7" :
         print ("Ok")
    else :
         print ("Periksa kembali pilihan anda!")


# Push

def Push (Stack):
  Push = input("Tambahkan item : ")       
  Stack.append (Push)
  print ("Item yang di tambahkan :",Stack[-1])
  Perintah ()


# Pop

def Pop (Stack) :
  if len(Stack) == 0 :
   print ("Stack Kosong")
  else :
   print ("Item yang diambil:",Stack.pop())
  Perintah ()


# Top

def Top (Stack):
  top = len (Stack) -1
  print ("Item teratas:" ,Stack[top])
  Perintah ()


# Size

def Size (Stack) :
  print ("Jumlah item:" ,len(Stack))
  Perintah ()


# Empty

def Empty (Stack) :
  if len(Stack) == 0 :
   print ("Stack Kosong")
  else :
   print ("Stack terisi")
  Perintah ()
  

#Tampilkan Stack

def Print (Stack):
  print ("Stack : ",Stack)
  Perintah ()

Perintah ()
```
