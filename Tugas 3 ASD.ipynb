```python

Item = ["Ayam","Sapi","Ikan"] 

#Menu

def Menu () :
    Menu = ["1. Tambah barang","2. Hapus barang", "3. Edit barang",
    "4. Menampilkan barang" , "5. Mengetahui barang", 
    "6. Menampilkan index"]
    for i in Menu :
        print(i)
    
    Opsi = input("Pilih menu : ")   
    if Opsi == "1" :
         Tambah ()
    elif Opsi == "2" :
         Hapus ()
    elif Opsi == "3" :
         Edit ()                  
    elif Opsi == "4" :
         Menampilkan ()
    elif Opsi == "5" :
         Cek () 
    elif Opsi == "6" :
         Index ()                          
         
#Tambah barang
        
def Tambah () :
     while True :
         Tambah = input("Tambahkan barang : ")       
         Item.append (Tambah)
         Pilihan = input("Ingin menambah barang lain? (Y/N) : ").upper ()
         if Pilihan == "Y" :
             pass 
         else :
             Menu ()    

#Hapus barang
             
def Hapus () :
    while True :
        Hapus = input("Hapus barang : ")
        Item.remove (Hapus)
        Pilihan = input("Ingin menghapus barang lagi? (Y/N) : ").upper ()
        if Pilihan == "Y" :
            pass 
        else :
             Menu ()                         

#Edit barang
               
def Edit () :
    while True :
        Edit = input("Barang yang ingin di ganti : ")   
        Ubah = input("Ganti ke : ")
        Item [Item.index(Edit)] = Ubah                                                                     
        Pilihan = input("Ingin mengganti barang lain? (Y/N) : ").upper ()
        if Pilihan == "Y" :
            pass 
        else :
            Menu () 
 
#Menampilkan Barang
                                   
def Menampilkan () :
    for i in Item :
        print(i)           
    Menampilkan = input("Tekan enter untuk kembali  : ")
    if Menampilkan == " " :
       Menu ()
    else :
       Menu ()                                                                                                                                                                         
 
#Cek barang
                                                                                                                                                                                   
def Cek () :  
    while True :
        Cek = input("Nama barang : ")                                                                                                                                                                                  
        if Cek in Item :
            print(Cek, "Barang tersedia!")                                                                                                                                                                                                   
            Pilihan = input("Ingin cek barang lain? (Y/N) : ").upper ()
            if Pilihan == "Y" :
                pass 
        else :
            Menu () 

#Menampilkan index barang

def Index () :
    while True :
        Index = input("Nama barang : ")                                                                                                                                                                                  
        if Index in Item :
            print(Index , "berada di index : ", Item.index(Index))                                                                                                                                                                                                   
            Pilihan = input("Ingin cek index barang yang lain? (Y/N) : ").upper ()
            if Pilihan == "Y" :
                pass 
        else :
            Menu ()
       
Menu ()
```
