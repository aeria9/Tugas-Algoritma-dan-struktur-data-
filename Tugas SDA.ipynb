```python
#Operator Aritmatika

Nilai= 5
Nilai2 = 3

#penjumlahan (+)
print(Nilai + Nilai2)

#pengurangan (-)
print(Nilai - Nilai2)

#perkalian (*)
print(Nilai * Nilai2)

#pembagian (/)
print(Nilai / Nilai2)

#sisa bagi (%)
print(Nilai % Nilai2)

#pangkat (**)
print(Nilai ** Nilai2)

#pembagian bulat (//)
print(Nilai // Nilai2)
```

```python
#Contoh penerapan Operator aritmatika

Nama = "Dandi Purnomo"
Nim = "D0221078"
Gajipokok = 1000000
Gajilembur = 5000
Waktulembur = 78
Pajak = 0.09

Totalpajak = int(Gajipokok * Pajak)
Gajikotor = int(Gajipokok - Totalpajak)
Gajibersih = int(Gajikotor + (Gajilembur * Waktulembur))

print("Nama : ",(Nama))
print("NIM : ",(Nim))
print("Gaji Pokok : ",(Gajipokok))
print("Gaji Lembur : ",(Gajilembur))
print("Waktu Lembur : ",(Waktulembur))
print("Pajak : ",(Pajak))
print("Total Pajak : ",(Totalpajak))
print("Gaji Kotor : ",(Gajikotor))
print("Gaji Bersih : ",(Gajibersih))
```

```python
#Operator Penugasan

#Sama dengan (=)
Nilai = 2
Nilai2 = 50

#Tambah sama dengan (+=)
#Nilai += 2 sama dengan Nilai = Nilai + 2
Nilai += 3

#Nilai =+ 3 sama dengan Nilai2 = 3
Nilai2 =+ 3
print(Nilai) 
print(Nilai2)

#Kurang sama dengan (-=)
Nilai -= 2
print(Nilai)

#Kali sama dengan (*=)
Nilai *= 8
print(Nilai)

#Bagi sama dengan (/=)
Nilai /= 2
print(Nilai)

#Sisa bagi sama dengan (%=)
Nilai %= 5
print(Nilai)

#Pangkat sama dengan (**=)
Nilai **= 6
print(Nilai)

#Pembagian bulat sama dengan (//=)
Nilai //= 4
print(Nilai)
```

```python
#Operator Perbandingan

A = 7
B = 5

#Sama dengan (==)
print(A == B)

#Tidak sama dengan (!=)
print(A != B)

#Lebih besar (>)
print(A > B)

#Lebih kecil (<)
print(A < B)

#Sama dengan atau lebih besar (>=)
print(A >= B)

#Sama dengan atau lebih kecil (<=)
print(A <= B)
```

```python
#Operator Logika

#And
print(True and True)
print(1 + 2 == 3 and True)

#Or
print(False or 1 > 5)
print(False or 5 > 2)

#Not
print(not(1 > 5))
print(not(1 < 5))
```
